```python
# Lets Calculate total amount spent for each campaign
total_spent = df.groupby("Campaign ID")["Amount Spent in INR"].sum()
```

```python
# to get the total spent amount
total_spent
```

```python
# Lets Define performance metrics
# The following are the features which can be considered as a performance metrics and these will make an impact in the resutls
performance_metrics = ["Reach", "Impressions", "Click-Through Rate (CTR)", "Unique Click-Through Rate (Unique CTR)"]
```

```python
# Lets Evaluate campaign performance based on our created performance metrics
performance_scores = df.groupby("Campaign ID")[performance_metrics].mean()
```

```python
# to get the performance scores
performance_scores
```

```python
# Lets Identify campaigns with high costs but relatively low performance
underperforming_campaigns = performance_scores[total_spent > 500]["Reach"].idxmin()
```

```python
# For Visualization - Compare relevant columns
comparison_columns = ["Reach", "Impressions", "Clicks", "Amount Spent in INR"]
comparison_data = df.groupby("Campaign ID")[comparison_columns].sum()
```

```python
# Plotting the comparison with values on top of bars
ax = comparison_data.plot(kind="bar", stacked=False, figsize=(15, 8))
plt.xlabel("Campaign ID")
plt.ylabel("Counts / Amount Spent")
plt.title("Comparison of Relevant Columns for Campaigns")
plt.show()
```

From the visual it is clear that the campaign 11 is has a poor performance with high cost.

```python
print("Recommendation: Discontinue the following campaign due to high costs and low performance:")
print(underperforming_campaigns)
```
